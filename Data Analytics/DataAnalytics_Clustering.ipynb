```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Income Data.csv')
```

```python
data.head(5)
```

```python
plt.figure(figsize = (5, 4), dpi = 105)
sns.scatterplot(x = 'age', y = 'income', data = data);
```

### Distance metrics to Find similarities

## KMeans

```python
from sklearn.cluster import KMeans
clusters = KMeans(n_clusters = 3)

clusters.fit(data)
```

```python
clusters.labels_
data['clusterid'] = clusters.labels_
```

```python
data.head(3)
```

```python
plt.figure(figsize = (5, 4), dpi = 105)
sns.scatterplot(x = 'age', y = 'income', data = data, hue = 'clusterid');
```

```python
markers = ['+', '^', '*']
sns.lmplot(x = 'age', y = 'income', data = data, hue = 'clusterid', markers = markers, fit_reg = False);
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
data_scaled = scaler.fit_transform(data[['age', 'income']])
```

```python
cluster_new = KMeans(n_clusters = 3)
cluster_new.fit(data_scaled)
```

```python
data['clusterid'] = cluster_new.labels_
```

```python
markers = ['+', '^', '*']
sns.lmplot(x = 'age', y = 'income', data = data, hue = 'clusterid', markers = markers, fit_reg = False);
```
