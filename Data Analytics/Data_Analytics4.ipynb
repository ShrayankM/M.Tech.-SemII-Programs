```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# 2 Sample T test

```python
df_yes = pd.read_excel('healthdrink.xlsx', 'healthdrink_yes')
df_no = pd.read_excel('healthdrink.xlsx', 'healthdrink_no')
```

```python
plt.figure(figsize = (5, 4), dpi = 125)
sns.distplot(df_yes, label = 'healthdrink_yes')
sns.distplot(df_no, label = 'healthdrink_no')
plt.legend();
```

```python
stats.ttest_ind(df_yes, df_no)
```

# Paired T-test


```python
paired_df = pd.read_csv('breakups.csv')
```

```python
paired_df.head(5)
```

```python
sns.distplot(paired_df['Before_Breakup'], label = 'before')
sns.distplot(paired_df['After_Breakup'], label = 'after')
plt.legend();
```

# Chi-Square Goodness of Fit test

```python
N = 500
C = 4
f_obs = [190, 185, 90, 35]
f_per = [35, 40, 20, 5]
f_exp = [i * 0.01 * 500 for i in f_per]
```

```python
f_exp
```

```python
stats.chisquare(f_obs, f_exp)
```
